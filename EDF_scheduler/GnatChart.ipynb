```python
import matplotlib.pyplot as plt

# Task execution intervals: (start_time, end_time)
tasks = {
    'T1': [(0, 200)],
    'T2': [(200, 300)],
    'T3': [(300, 400)],
    'T4': [(400, 600)]
}

fig, ax = plt.subplots(figsize=(10, 4))

# Assign Y-axis positions
task_y = {'T1': 30, 'T2': 20, 'T3': 10, 'T4': 0}
colors = {'T1': 'skyblue', 'T2': 'lightgreen', 'T3': 'lightcoral', 'T4': 'gold'}

for task, intervals in tasks.items():
    for start, end in intervals:
        ax.broken_barh([(start, end - start)], (task_y[task], 8), facecolors=colors[task])
        ax.text(start + 5, task_y[task] + 1, f"{task}", fontsize=9, color='black')

# Format chart
ax.set_ylim(-5, 40)
ax.set_xlim(0, 700)
ax.set_xlabel("Time (ms)")
ax.set_yticks([4, 14, 24, 34])
ax.set_yticklabels(['T4', 'T3', 'T2', 'T1'])
ax.grid(True)
plt.title("Gantt Chart of EDF Task Scheduling")
plt.show()
```
