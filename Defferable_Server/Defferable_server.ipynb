```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Define the tasks with (start_time, duration, label, color, row_position)
tasks = [
    (0, 5, 'Periodic A', 'skyblue', 3),
    (5, 5, 'Periodic B', 'lightgreen', 2),
    (12, 2, 'DS: Handles Task X', 'orange', 1),
    (20, 5, 'Periodic A', 'skyblue', 3),
    (25, 5, 'Periodic B', 'lightgreen', 2),
    (27, 2, 'DS: Handles Task Y', 'orange', 1),
]

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 4))

# Draw each task block with offset label
for start, duration, label, color, row in tasks:
    ax.broken_barh([(start, duration)], (row - 0.4, 0.8), facecolors=color)
    ax.text(start + 0.2, row + 0.2, label, va='bottom', ha='left', fontsize=8, color='black')

# Y-axis setup
ax.set_ylim(0.5, 3.8)
ax.set_yticks([1, 2, 3])
ax.set_yticklabels(['Deferrable Server (DS)', 'Periodic Task B', 'Periodic Task A'])

# X-axis setup
ax.set_xlim(0, 30)
ax.set_xticks(range(0, 31, 5))
ax.set_xlabel('Time (ms)')
ax.set_title('Gantt Chart: How the Deferrable Server Responds to Aperiodic Tasks')
ax.grid(True)

# Legend outside the plot
legend_patches = [
    mpatches.Patch(color='skyblue', label='Periodic Task A'),
    mpatches.Patch(color='lightgreen', label='Periodic Task B'),
    mpatches.Patch(color='orange', label='Deferrable Server Execution'),
]
ax.legend(handles=legend_patches, bbox_to_anchor=(1.01, 1), loc='upper left')

# Save and show the plot
plt.tight_layout()
plt.savefig("deferrable_server_gantt_matched.png", bbox_inches='tight')
plt.show()
```
